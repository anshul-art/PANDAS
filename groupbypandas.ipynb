<h1> Group by and Aggregating</h1>

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\dell\Downloads\vgsales.csv")
df
```

```python
group_by = df.groupby('Genre')                                                           
                      
                      
                      
  
print(group_by)
```

```python
df.groupby('Genre')['Global_Sales'].sum()                    
  
```

```python
df.groupby('Genre')['Global_Sales'].max()
```

```python
df.groupby('Genre')['Global_Sales'].count()
```

```python
df.groupby('Genre').agg({'Global_Sales':['mean','max','count','sum']})                                          
```

```python
df.groupby('Genre').agg({'Global_Sales':['mean','max','count','sum'],'JP_Sales':['mean','max','count','sum']})
```

```python
df.groupby('Genre').describe()
```
