<h1>Merge,Join and concatenate</h1>

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\dell\Downloads\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\dell\Downloads\LOTR 2.csv")                    
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how='inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how='inner', on = ['FellowshipID','FirstName'])
```

```python
df1.merge(df2, how='left') #same values from right and all values from left
```

```python
df1.merge(df2, how='right')#same values from left and all values from right
```

```python
df1.merge(df2, how='outer')#all the values from both left and right except duplicate ones
```

<h5>USE MERGE,WHEN WORKING WITH COLUMNS</h5>

<h5>USE JOIN,WHEN WORKING WITH INDEXES</h5>

```python
pd.concat([df1,df2]) #combines the dfs
```
